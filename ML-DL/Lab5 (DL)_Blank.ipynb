# SVR

# Importing the libraries

```python
import pandas as pd
import numpy as np
```

# Read World bank CSV

```python
df = pd.read_csv('WorldBank.csv')
df.head()
```

# Perform conditional selection to find - Population ages 15-64 (% of total population)

```python
df1 = df[df['Indicator Name'] == 'Population ages 15-64 (% of total population)']
df1
```

# Divide the data into input and output

```python
X = np.arange(1960, 2020).reshape(-1,1)
y = df1.values[0][4:-1].reshape(-1,1)
y
```

```python
import matplotlib.pyplot as plt

plt.scatter(X,y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_X = StandardScaler()
sc_y = StandardScaler()

x_scaled = sc_X.fit_transform(X)
y_scaled = sc_y.fit_transform(y)

y_scaled
```

# Plot scatter plot of Population ages 15-64 (% of total population)

# Feature Scaling (Mandatory for SVR)**
SVR is highly sensitive to the range of data points. If we don't scale (normalize) the data, the model will fail to find the correct hyperplane.

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.2)
```

# Fitting SVR on 3 Different Kernel on dataset

```python
from sklearn.svm import SVR

model_1 = SVR()

model_1.fit(X_train, y_train.ravel())
```

```python
lin = SVR(kernel='linear')
poly = SVR(kernel='poly', degree=3)
rbf = SVR(kernel='rbf')

lin.fit(X_train, y_train.ravel())
poly.fit(X_train, y_train.ravel())
rbf.fit(X_train, y_train.ravel())
```

```python
X_grid = np.arange(min(X_test), max(X_test), 1)
X_grid = X_grid.reshape((len(X_grid), 1))
X_grid
```

```python
plt.scatter(X,y, color='red', label='delta')

plt.plot(sc_X.inverse_transform(X_grid), sc_y.inverse_transform(lin.predict(X_grid).reshape(-1,1)), color='blue')
plt.plot(sc_X.inverse_transform(X_grid), sc_y.inverse_transform(poly.predict(X_grid).reshape(-1,1)), color='orange')
plt.plot(sc_X.inverse_transform(X_grid), sc_y.inverse_transform(rbf.predict(X_grid).reshape(-1,1)), color='green')

plt.show()
```

# Predict the x_test using 3 Kernel

```python
lin.score(X_train,y_train)
```

```python
rbf.score(X_test,y_test)
```
